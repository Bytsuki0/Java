```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline

caminho_arquivo = r"D:\Code\Java\threads.csv"
if not os.path.isfile(caminho_arquivo):
    raise FileNotFoundError(f"Arquivo não encontrado: {caminho_arquivo}")

# 1) Leitura inicial sem conversão
df = pd.read_csv(caminho_arquivo, header=0, dtype=str)

# 2) Lista das colunas que precisam de conversão
colunas_numericas = ['dezmil', 'cemmil', 'milhao', 'dezmilhao', 'cemmilhao']

# 3) Para cada coluna:
for col in colunas_numericas:
    # a) substitui vírgula por ponto
    df[col] = df[col].str.replace(',', '.', regex=False)
    # b) transforma a string em float
    df[col] = df[col].astype(float)

# 4) Se a coluna 'Threads' também estiver como string, converta-a:
df['Threads'] = df['Threads'].astype(int)  # ou float, se for o caso

# 5) Agora você pode plotar:
style.use('ggplot')
plt.plot(df['Threads'], df['dezmil'], marker='o', linestyle='-',color='r',  label='Dez mil Execuções')
plt.plot(df['Threads'], df['cemmil'], marker='o', linestyle='-',color='b',  label='Cem mil Execuções')
plt.title("Tempo de Execução vs Número de Threads (menos de um milhão de execuções)")
plt.xlabel("Número de Threads")
plt.ylabel("Tempo de Execução (s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

style.use('ggplot')
plt.plot(df['Threads'], df['milhao'], marker='o', linestyle='-',color='y',  label='Milhão de Execuções')
plt.title("Tempo de Execução vs Número de Threads")
plt.xlabel("Número de Threads")
plt.ylabel("Tempo de Execução (s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
style.use('ggplot')
plt.plot(df['Threads'], df['dezmilhao'], marker='o', linestyle='-',color='r',  label='Dez milhões de Execuções')
plt.title("Tempo de Execução vs Número de Threads")
plt.xlabel("Número de Threads")
plt.ylabel("Tempo de Execução (s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
style.use('ggplot')
plt.plot(df['Threads'], df['cemmilhao'], marker='o', linestyle='-',color='g',  label='Cem milhões de Execuções')
plt.title("Tempo de Execução vs Número de Threads")
plt.xlabel("Número de Threads")
plt.ylabel("Tempo de Execução (s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

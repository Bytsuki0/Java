```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline

caminho_arquivo = r"D:\Code\Java\threads.csv"
if not os.path.isfile(caminho_arquivo):
    raise FileNotFoundError(f"Arquivo no encontrado: {caminho_arquivo}")

# 1) Leitura inicial sem converso
df = pd.read_csv(caminho_arquivo, header=0, dtype=str)

# 2) Lista das colunas que precisam de converso
colunas_numericas = ['dezmil', 'cemmil', 'milhao', 'dezmilhao', 'cemmilhao']

# 3) Para cada coluna:
for col in colunas_numericas:
    # a) substitui vrgula por ponto
    df[col] = df[col].str.replace(',', '.', regex=False)
    # b) transforma a string em float
    df[col] = df[col].astype(float)

# 4) Se a coluna 'Threads' tambm estiver como string, converta-a:
df['Threads'] = df['Threads'].astype(int)  # ou float, se for o caso

# 5) Agora voc pode plotar:
style.use('ggplot')
plt.plot(df['Threads'], df['dezmil'], marker='o', linestyle='-',color='r',  label='Dez mil Execues')
plt.plot(df['Threads'], df['cemmil'], marker='o', linestyle='-',color='b',  label='Cem mil Execues')
plt.title("Tempo de Execuo vs Nmero de Threads (menos de um milho de execues)")
plt.xlabel("Nmero de Threads")
plt.ylabel("Tempo de Execuo (s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

style.use('ggplot')
plt.plot(df['Threads'], df['milhao'], marker='o', linestyle='-',color='y',  label='Milho de Execues')
plt.title("Tempo de Execuo vs Nmero de Threads")
plt.xlabel("Nmero de Threads")
plt.ylabel("Tempo de Execuo (s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
style.use('ggplot')
plt.plot(df['Threads'], df['dezmilhao'], marker='o', linestyle='-',color='r',  label='Dez milhes de Execues')
plt.title("Tempo de Execuo vs Nmero de Threads")
plt.xlabel("Nmero de Threads")
plt.ylabel("Tempo de Execuo (s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
style.use('ggplot')
plt.plot(df['Threads'], df['cemmilhao'], marker='o', linestyle='-',color='g',  label='Cem milhes de Execues')
plt.title("Tempo de Execuo vs Nmero de Threads")
plt.xlabel("Nmero de Threads")
plt.ylabel("Tempo de Execuo (s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
